```python
# Determine number of fraud cases in dataset
fraud = data[data['Dependent_count'] == 1]
valid = data[data['Dependent_count'] == 0]
outlierFraction = len(fraud)/float(len(valid))
print(outlierFraction)
print('Fraud Cases: {}'.format(len(data[data['Dependent_count'] == 1])))
print('Valid Transactions: {}'.format(len(data[data['Dependent_count'] == 0])))
```
